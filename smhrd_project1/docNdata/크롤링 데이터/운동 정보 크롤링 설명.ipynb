```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
from bs4 import BeautifulSoup as bs

import csv
import os # 파일시스템을 위한 라이브러리 
           # ex)파일, 폴더를 생성, 삭제, 존재여부 파악
from urllib.request import urlretrieve as urlre # 이미지경로를 파일로 저장
```

```python
# 이름, 정보, 부위, 장비, 강도 정보를 담기위한 리스트 생성
name_list = []    
info_list = []
rank_list = []
part_list = []
equipment_list =[]
important_list =[]


driver = wb.Chrome()

#덤벨로 하는 운동 정보 가져오기
for j in range(1,5): # 4페이지 까지 크롤링
    url = f'https://www.acefitness.org/resources/everyone/exercise-library/equipment/dumbbells/?page={j}'
    driver.get(url)
    driver.maximize_window() 
    for i in range(12): # 한 페이지 안에 12개의 운동 정보가 있습니다.
        try:
            img = driver.find_elements_by_css_selector('.exercise-card-grid__cell') # 이미지 태그 클릭해서
            img[i].click()
            time.sleep(0.5)
        except:
            break
            time.sleep(0.5)
        # 이름, 정보 ,부위, 장비별, 강도  정보를 가져옴
        soup = bs(driver.page_source, 'lxml')
        name = soup.select('body > div.content > header > div > div.exercise-hero__content-wrap > h1')
        info = soup.select('body > div.content > div.content__wrap > div > main > div.exercise-post > section > div.exercise-post__step-content')
        part = soup.select('body > div.content > header > div > div.exercise-hero__content-wrap > div > dl > div.exercise-info__term.exercise-info__term--body-part > dd')
        equipment = soup.select('body > div.content > header > div > div.exercise-hero__content-wrap > div > dl > div.exercise-info__term.exercise-info__term--equipment > dd')
        important = soup.select('body > div.content > header > div > div.exercise-hero__content-wrap > div > dl > div.exercise-info__term.exercise-info__term--difficulty > dd > div > span')
        
        # 리스트에 정보들 저장
        name_list.append(name[0].text.strip())
        info_list.append(info[0].text.strip())
        part_list.append(part[0].text.strip())
        equipment_list.append(equipment[0].text.strip())
        important_list.append(important[0].text.strip())
        rank_list.append(i+1)
        
        driver.back()
        time.sleep(5)
```
